```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('walmart_train.csv')
feature=pd.read_csv('walmart_features.csv')
store=pd.read_csv('walmart_stores.csv')
test=pd.read_csv('walmart_test.csv')
```

```python
train=train.merge(feature,how='left')
```

```python
train=train.merge(store,how='left')
train
```

```python
for mark in train.columns[7:12]:
    train[mark+'_zero']=train[mark].fillna(0)
```

```python
for mark in train.columns[7:12]:
    mean=train[mark].mean()
    train[mark+'_mean']=train[mark].fillna(mean)
```

```python
for mark in train.columns[7:12]:
    median=train[mark].median()
    train[mark+'_median']=train[mark].fillna(median)
```

```python
for mark in train.columns[7:12]:
    mode=train[mark].mode()
    train[mark+'_mode']=train[mark].fillna(mode)
```

```python
for mark in train.columns[7:12]:
    mode=train[mark].mode()
    train[mark+'_mode']=train[mark].fillna(mode)
```

```python
for mark in train.columns[7:12]:
    extr_outlier=train[mark].mean()+3*train[mark].std()
    train[mark+'ext_outlier']=train[mark].fillna(extr_outlier)
```

```python
for mark in train.columns[7:12]:
    lowe_outlier=train[mark].mean()-3*train[mark].std()
    train[mark+'low_outlier']=train[mark].fillna(lowe_outlier)
```
